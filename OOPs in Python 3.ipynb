## ***Encapsulation***(hiding of data)

Encapsulation is one of the fundamental concepts in object-oriented programming (OOP).This puts restrictions on accessing variables and methods directly and can prevent the accidental modification of data.an object’s variable can only be changed by an object’s method. Those types of variables are known as private variables.

Some time it is very important to make some data and Method private so that we limit there use and and change in it.

Unlike java there is no key word like public ,private and protected here we use double underScore for making variable or method private

                       self.__show="Hiii"     ----> private variable
                       self.__list=[1,2,3]     ----->private variable





```python
class xyz:
  def __init__(self):
    self.a=20                #public variable
    self._b=100              #protected variable
    self.__c=50              #privat variable

obj=xyz()
print(obj.a)
print(obj._b)
# print(obj__c) it give error
```

But in python nothing is completly private if you can do chages in private variable and method you can do it  by using getter and setter method.

```python
class A:
  def __init__(self):
    self.var1=20
    self.__var2=30

  def set_var2(self,value):
    self.__var2=value
  def get_var2(self):
    return self.__var2

obj1=A()
obj1.set_var2(11)
print(obj1.get_var2())
```

```python
# private method:
class AA:
  def __Show(self):
    print("Hii i am a private method")

  def get_show(self):
    self.__Show()

obj2=AA()
obj2.get_show()
```

# ***ABSTRACTION***

Abstraction is used to hide the internal functionality of the function from the users. The users only interact with the basic implementation of the function, but inner working is hidden. User is familiar with that "what function does" but they don't know "how it does."

In Python, an abstraction is used to hide the irrelevant data/class in order to reduce the complexity. It also enhances the application efficiency.



```python
from abc import ABC ,abstractmethod
class shape(ABC):
  @abstractmethod
  def area(self):
    pass

  def show(self):
    print("Hiii")

class circle(shape):
  def __init__(self,S):
    self.side=S

  def area(self):
    return self.side*self.side

onj1=circle(5)
onj1.area()
```
