# ***Inheritance***



>

```
*   It is a Mechanism Thats allows you to create a hierarchy of classes that share properties and method to derived class
*   It allows us to create another class that inharit all properties of other class


```





````python
class A:                                      # parent class or Base class
  def feature(self):
    print("This is Feature")

class B(A):                                   # child class or derived class
  def feature1(self):
    print("This is Feature 1")

#note class B inharit all properties of class A as a is a parent class

obj1=B()                                     #object of child class
obj1.feature()                               # we are accessing parents method with child object
````

````python
# Single inharitance:- when a class is derived from a single parent class is known as single inheriatance

class C:
  def Feature2(self):
    print("Welcome in feature 2")

class D(C):
  def feature3(self):
    print("Welcome in feature 3")

object=D()
object.Feature2()
object.feature3()
````

````python
# Multipal inhariatance :- when a class is derived from two or more then two classes is known as multiple inhariatance in simple word a
# having more the one parent
class mother:
  def Dna(self):
    print("Hii son")
class Father:
  def geans(self):
    print("Hello my boy")

class child(mother, Father):
  def gardian(self):
    print("Welcome in multiple inhariatance")

obj2=child()
obj2.Dna()
obj2.geans()
obj2.gardian()



````

````python
#Multi-Level in hharitance:-In multilevel inheritance, features of the base class and the derived class are further inherited into the new derived class.
#This is similar to a relationship representing a child and a grandfather.

class one:
  def quality1(self):
    print("You have one quality")
class Two(one):
  def quality2(self):
    print("You have two quality")
class three(Two):
  def quality3(self):
    print("It has Three quality")
objec1=three()
objec1.quality1()
objec1.quality2()
objec1.quality3()
````

**Behaviour of constructor**

if parent and child both have constructor the if we create object of child class constructor of child class will run .


````python
class AAA:
  def __init__(self):
    print("in A init")

class BBB(AAA):
  def __init__(self):
    print("in B init")

ojc=BBB()
````

But  if we want that constructor of parent class will run first then child class constructor run then we can use super()

````python
class A1:
  def __init__(self):
    print("in A1 init")
class B1(A1):
  def __init__(self):
    super().__init__()
    print("in B1 init")

chij=B1()
````

MRO(Method Resolution order) in case of constructor

````python
class Aone:
  def __init__(self):
    print("In Aone")
class Btwo:
  def __init__(self):
    print("In Btwo")

class Cthree(Aone,Btwo):
  def __init__(self):
    super().__init__()
    print("welcom in Cthree")

ogj=Cthree()
````

# *`Duck Typing`*

Duck Typing is a term commonly related to dynamically typed programming languages and polymorphism. The idea behind this principle is that the code itself does not care about whether an object is a duck, but instead it does only care about whether it quacks.

If it walks like a duck and queks like a duck then it is a duck

````python
class Duck:
  def walk(self):
    print("Thapak thapak thapak")
class Horse:
  def walk(self):
    print("Tabdak tabdak tabdak")

def myfuntion(obj):
  obj.walk()

A=Horse()                    # It is a horse but bcz like duck its also executes on calling so it should be considered as duck
myfuntion(A)
````

# Method overloading in python

when two or more mathod have different parameter or different type of parameter
or both these method are called over loading method unlike C++ Python doesn't support overloading but we can achive it in another way

note:-The problem with method overloading in Python is that we may overload the methods but can only use the latest defined method.

````python
def sum(a,b):
  print(a+b)

def sum(a,b,c):
  print(a+b+c)

#sum(1,2)  it gives error
sum(1,2,3) as is latest defined funtion its work properly
````

We can achive Method Overloading  by Multiple Dispatch Decorator

````python

from multipledispatch import dispatch

@dispatch(int,int)
def product(first ,second):
  return first*second

@dispatch(int,int,int)
def product(first, second, third):
  return first*second*third

print(product(1,4))
print(product(2,3,4))
````

# ***Method Overriding***
When a method in a subclass has the same name, same parameters or signature and same return type(or sub-type) as a method in its super-class, then the method in the subclass is said to override the method in the super-class.


````python
class parent:
  def show(self):
    print("Hii You are in parent class")

class child(parent):
  def show(self):
    print("Hii You are in child class")

obj12=child()

obj12.show()
````

# ***Operator Overloading***

Operator Overloading means giving extended meaning beyond their predefined operational meaning. For example operator + is used to add two integers as well as join two strings and merge two lists. It is achievable because ‘+’ operator is overloaded by int class and str class. You might have noticed that the same built-in operator or function shows different behavior for objects of different classes, this is called Operator Overloading.



````python
print(1+2)
print("Hii"+" Riya")
print([1,2,3,4]+[5,6,7,8])
````
