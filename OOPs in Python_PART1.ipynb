````python
#Instant Method
class C:
  def __init__(self, phone, email):
    self.phone=phone
    self.email=email
  def show(self):                                            #instant Method
    print(f"Phone number of person{self.phone}")
    print(f"Email of the person is {self.email}")
  def set_phone(self, Value):                                #Mutator Method:-only for set value
    self.phone=Value
  def get_phone(self,Value):                                 #Accessor Method:- Only for get value
    return self.phone


per4=C("7283032588","aec.cse.riyakumari@gmail.com")
per4.show()
per4.set_phone("8340445078")
per4.show()





````

````python
#class Method
class C:
  def __init__(self, code, id):
    self.code=code
    self.id=id
  def show1(self):
    print(f"Phone number of person{self.code}")
    print(f"Email of the person is {self.id}")
  @classmethod
  def fromstr(cls):
    print("Hiii")


per9=C(123, 1234)
C.fromstr()


````

````python
#Static Method

class animal:
  def __init__(self,Gender):
    self.Gender=Gender

  @staticmethod
  def Check(gens):
    if gens=="male":
      print("boy")
    else:
      print("girl")
var1=animal("male")
animal.Check("male")
var1.Check("male")
````

**Note**


*   A static method is method in class that doesn't need implicit first argument i.e (self)
*   This method is  access with help of class and object both but you cannot change anything in class state






**Class method vs Static Method**

*   A class method takes cls as the first parameter while a static method needs no specific parameters.
*   A class method can access or modify the class state while a static method canâ€™t access or modify it.
*   In general, static methods know nothing about the class state. They are utility-type methods that take some parameters and work upon those parameters. On the other hand class methods must have class as a parameter.
*   We use @classmethod decorator in python to create a class method and we use @staticmethod decorator to create a static method in python.
